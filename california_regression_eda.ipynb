```python
corr_matriz=df.corr()

corr_matriz.head(10)
```

```python
df.hist(bins=50, figsize=(15,10))  
plt.suptitle("Distribucin de variables - California Housing", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])  
plt.show()
```

```python
df["AveRooms"].hist(bins=50, figsize=(6,4), range=(0,20))
plt.title("AveRooms (zoom 020)")
plt.show()
```

```python
df["AveOccup"].hist(bins=50, figsize=(6,4), range=(0,20))
plt.title("AveOccup (zoom 020)")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr_matriz,annot=True,cmap="coolwarm",center=0)
plt.title("matriz de correlacion - California Housing")
plt.show()
```

```python
variables = ["MedInc", "HouseAge", "AveRooms", "AveBedrms", "Population", "AveOccup", "Latitude", "Longitude"]

for var in variables:
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=var, y="MedHouseVal", data=df, alpha=0.3)
    plt.title(f"{var} vs MedHouseVal")
    plt.show()
```

```python
variables = ["AveRooms", "AveBedrms", "Population", "AveOccup"]

plt.figure(figsize=(12,6))
for i, var in enumerate(variables, 1):
    plt.subplot(1, 4, i)
    sns.boxplot(y=df[var])
    plt.title(var)

plt.tight_layout()
plt.show()
```
